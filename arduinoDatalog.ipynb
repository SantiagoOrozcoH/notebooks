# Data processing - Temperature and Humidity from Arduino dataLogger 

```python
import matplotlib.pyplot as plt
import pandas as pd
import requests
import json
```

```python
url = "http://api2.orozcoh.com:4000/v1/data/get-all"
res = requests.get(url)

print(res.status_code)
if res.status_code == 200:
  print("data ok")
  data = res.text
  data_j = json.loads(data)
  df_data = pd.DataFrame(data_j)
```

```python
print(df_data.head())
```

```python
time = df_data["timestamp"].to_numpy()
temp = df_data["Temp (°C)"].to_numpy()
humidity = df_data["Humidity (%)"].to_numpy()
light = df_data["Light"].to_numpy()
```

```python
plt.plot(time, temp)
```

```python
plt.plot(time, humidity)
```

```python
plt.plot(time, light)
```
